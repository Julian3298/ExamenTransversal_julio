```python
from itertools import cycle
from datetime import date

nro_pisos = 10
depa_x_piso = 4
depa_vendido = []
compradores = []

def digito_verificador(run):
  reversed_digits = map(int, reversed(str(run)))
  factors = cycle(range(2, 8))
  s = sum(d * f for d, f in zip(reversed_digits, factors))
  return (-s) % 11

def validar_run(run):
    rut = str(run) + "-" + str(digito_verificador(run))
    return run


def mostrar_menu():
    print("\n******* MENÚ CASA FELIZ*******\n")
    print("1. Comprar Departamento")
    print("2. Mostrar Departamentos Disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")


def comprar_departamento():
    print("\n---DEPARTAMENTOS DISPONIBLES---")
    print("PISO\tTIPO\tESTADO")
    for piso in range(1, nro_pisos + 1):
        for letra in ['A', 'B', 'C', 'D']:
            depa = letra + str(piso)
            if depa in depa_vendido:
                std = "Vendido"
            else:
                std = "Disponible"
            print(f"{piso}\t{depa}\t{std}")

    depa = input("Ingrese el departamento a comprar (ejemplo: A1): ")
    if depa in depa_vendido:
        print("El departamento seleccionado no está disponible.")
    else:
        t_depa = depa[0]
        run = input("Ingrese el RUN del comprador sin punto ni guion ni digito verificador: ")
        digito= int(input("Ingrese digito verificador "))
        if digito_verificador(run)==digito:
          if depa[1:].isdigit():
              piso = int(depa[1:])
              if 1 <= piso <= nro_pisos and t_depa in ['A', 'B', 'C', 'D']:
                  if t_depa == 'A':
                      precio = 3800
                  elif t_depa == 'B':
                      precio = 3000
                  elif t_depa == 'C':
                      precio = 2800
                  elif t_depa == 'D':
                      precio = 3500
                  depa_vendido.append(depa)
                  compradores.append((run, depa, "Compra"))
                  print("\nLa operación ha sido un exito!.\n")
              else:
                  print("\nEl departamento ingresado no existe.")
          else:
              print("\nDepartamento inválido.")
        else:
          print("\nEL RUN ingresado es invalido, ingreselo nuevamente.")
          return

def mostrar_depas_disponibles():
    print("\n---DEPARTAMENTOS DISPONIBLES---")
    print("PISO\t TIPO\t ESTADO")
    for piso in range(1, nro_pisos + 1):
        for letra in ['A', 'B', 'C', 'D']:
            depa = letra + str(piso)
            if depa in depa_vendido:
                std = "Vendido"
            else:
                std = "Disponible"
            print(f"{piso}\t{depa}\t{std}")


def lista_compradores():
    if len(compradores) == 0:
        print("No se encontraron compradores registrados.")
    else:
        compradores_ordenados = sorted(compradores, key=lambda x: x[0])
        print("RUN\t\tDEPARTAMENTO\tTIPO")
        for comprador in compradores_ordenados:
            print(f"{comprador[0]}\t{comprador[1]}\t\t{comprador[2]}")


def mostrar_ganancias():
  suma_total = 0
  total_cant = 0
  precios = {
        'A': {'venta': 3800},
        'B': {'venta': 3000},
        'C': {'venta': 2800},
        'D': {'venta': 3500},
  }

  ventas_totales = {}
  for depa in depa_vendido:
      tipo = depa[0]
      if tipo not in ventas_totales:
          ventas_totales[tipo] = 1
      else:
          ventas_totales[tipo] += 1

  print("TIPO DE DEPARTAMENTO\t       CANTIDAD\t      TOTAL")
  for tipo, cantidad in ventas_totales.items():
      precio_venta = precios[tipo]['venta']
      total = precio_venta * cantidad
      print(f"TIPO {tipo}  {precios[tipo]['venta']} UF\t\t\t{cantidad}\t      {total} UF")
      suma_total = suma_total + total
      total_cant = total_cant + cantidad
  print(f"TOTAL                           {total_cant}\t      {suma_total} UF" )

def menu():
    print("\n*****Bienvenido a la Inmobiliaria 'CASA FELIZ'*****")

    while True:
        mostrar_menu()
        opc = input("Seleccione una opción: ")

        if opc == '1':
            comprar_departamento()
        elif opc == '2':
            mostrar_depas_disponibles()
        elif opc == '3':
            lista_compradores()
        elif opc == '4':
            mostrar_ganancias()
        elif opc == '5':
            print("---GRACIAS POR UTILIZAR ESTE SISTEMA---")
            print("---ALUMNO JULIO SILVA TORRES, UN PLACER---")
            print("FECHA: " + str(date.today()))

            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")
menu()
```
